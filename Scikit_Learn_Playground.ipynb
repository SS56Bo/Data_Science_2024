```python
import numpy as np
import matplotlib.pyplot as mtlb
import seaborn as sns
import math
import pandas as pd
```

```python
N=40
x=np.arange(N)
y=4*x-math.log(N,8) + (9*np.random.randn(N))
```

```python
df = pd.DataFrame({'X:':x, "Y:":y})
```

```python
mtlb.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(x[:,np.newaxis],y)
yfit = model.predict(x[:,np.newaxis])
mtlb.scatter(x,y)
mtlb.plot(x,yfit,color='red')
mtlb.grid(visible=None)
```

```python
from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=500, centers=3, cluster_std=1.2)
mtlb.scatter(X[:,0],X[:,1],c=y, cmap='autumn')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33)
```

```python
from sklearn.svm import SVC
```

```python
mdl =SVC(gamma='auto')
mdl.fit(X_train, y_train)
```

```python
y_predict=mdl.predict(X_test)
```

```python
acc=100*np.sum(y_predict==y_test)/y_test.size
print("Accuracy is :",acc,"%")
```

```python
from sklearn.ensemble import RandomForestClassifier
mdl = RandomForestClassifier(n_estimators=20)
mdl.fit(X_train, y_train)
```

```python
y_predict=mdl.predict(X_test)
acc=100*np.sum(y_predict==y_test)/y_test.size
print("Accuracy is :",acc,"%")
```
