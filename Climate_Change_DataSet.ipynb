```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mtl
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv('climate_change_indicators.csv')
```

```python
df.drop(['ObjectId'],axis=1,inplace=True)
```

```python
df.drop(['Indicator','Source','Unit','CTS_Code','CTS_Name','CTS_Full_Descriptor'],axis=1,inplace=True)
```

```python
df.rename(columns={'F1961':'Data - 1961','F1962':'Data - 1962','F1963':'Data - 1963'},inplace=True)
```

```python
df2=df[['Country','Data - 1961','Data - 1962','Data - 1963']]
```

```python
df2=df2.fillna(0)
```

```python
data = pd.DataFrame({'Countries':df2.Country,'Data - 1961':df2['Data - 1961'],'Data - 1962':df2['Data - 1962'],'Data - 1963':df2['Data - 1963']})
```

```python
for col in data.columns:
  mtl.hist(data[col],density=True, alpha=0.5)
```
