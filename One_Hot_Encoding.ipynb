```python
import pandas as pd
import matplotlib.pyplot as mtp
import numpy as np
```

```python
data = pd.read_csv('/content/adult.data', header=None)
```

Performing One Hot Encoding


```python
encoded_Data = pd.get_dummies(data, columns=[1,3,5,6,7,8,9,13])
```

```python
encoded_Data.head()
```

```python
y=encoded_Data.loc[:,14]
```

```python
val = np.unique(y)
```

```python
idx = encoded_Data.loc[:,14]==val[0]
```

```python
encoded_Data.loc[idx,:]=0
encoded_Data.loc[~idx,:]=1
```

```python
np.unique(encoded_Data.loc[:,14])
```

```python
encoded_Data.to_csv("updated_data.csv")
```
